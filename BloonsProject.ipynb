```python
#Import libraries
import pandas as pd
import re
```

```python
# Webscraping

url = "https://bloons.fandom.com/wiki/Tower_Price_Lists"
tables = pd.read_html(url, header=0)  # returns a list of DataFrame objects

print(f"Number of tables found: {len(tables)}")

# Keep only tables 2 to 5
t2, t3, t4, t5 = tables[2:6]

# Combine them using t2's column headers
combined_df = pd.concat([t2, t3, t4, t5], ignore_index=True)

#Drop unused data
combined_df = combined_df.drop(columns=['Total Costs'])
combined_df = combined_df.drop(columns=['Unnamed: 7'])

#Removes * from dataset (became a problem in the next step)
combined_df['Upgrade 5'] = combined_df['Upgrade 5'].str.replace("*",'')

# Print to Excel File
# combined_df.to_excel(r"C:\Users\wills\Downloads\tower_prices.xlsx", index=False)

combined_df.head()

```

Now we need to seperate each row into multiple rows of data using the information


```python
#Copy of dataset
df2 = combined_df.copy()
# Regex formula for extracting data from each row. Regex Formula was pulled https://docs.python.org/3/library/re.html

# Tower Conversion
df2[['Base Tower Name', 'Easy Base Tower Cost', 'Medium Base Tower Cost', 'Hard Base Tower Cost', 'Impoppable Base Tower Cost']] = df2['Tower'].str.extract(
    r'^(.*?)\s+E\$(\d+)\s+M\$(\d+)\s+H\$(\d+)\s+I\$(\d+)$')
df2 = df2.drop(columns="Tower")

# Extract Tier 1 upgrade data
df2[['Easy Tier 1 Cost', 'Medium Tier 1 Cost', 'Hard Tier 1 Cost', 'Impoppable Tier 1 Cost']] = df2['Upgrade 1'].str.extract(
    r'E\$(\d+)\s+M\$(\d+)\s+H\$(\d+)\s+I\$(\d+)$')
df2 = df2.drop(columns="Upgrade 1")

# Extract Tier 2 upgrade data
df2[['Easy Tier 2 Cost', 'Medium Tier 2 Cost', 'Hard Tier 2 Cost', 'Impoppable Tier 2 Cost']] = df2['Upgrade 2'].str.extract(
    r'E\$(\d+)\s+M\$(\d+)\s+H\$(\d+)\s+I\$(\d+)$')
df2 = df2.drop(columns="Upgrade 2")

# Extract Tier 3 upgrade data
df2[['Easy Tier 3 Cost', 'Medium Tier 3 Cost', 'Hard Tier 3 Cost', 'Impoppable Tier 3 Cost']] = df2['Upgrade 3'].str.extract(
    r'E\$(\d+)\s+M\$(\d+)\s+H\$(\d+)\s+I\$(\d+)$')
df2 = df2.drop(columns="Upgrade 3")

# Extract Tier 4 upgrade data
df2[['Easy Tier 4 Cost', 'Medium Tier 4 Cost', 'Hard Tier 4 Cost', 'Impoppable Tier 4 Cost']] = df2['Upgrade 4'].str.extract(
    r'E\$([\d,]+)\s+M\$([\d,]+)\s+H\$([\d,]+)\s+I\$([\d,]+)$')
df2 = df2.drop(columns="Upgrade 4")

# Extract Tier 5 upgrade data
df2[['Easy Tier 5 Cost', 'Medium Tier 5 Cost', 'Hard Tier 5 Cost', 'Impoppable Tier 5 Cost']] = df2['Upgrade 5'].str.extract(
    r'E\$([\d,]+)\s+M\$([\d,]+)\s+H\$([\d,]+)\s+I\$([\d,]+)$')

df2 = df2.drop(columns="Upgrade 5")

#df2.to_excel(r"C:\Users\wills\Downloads\tower_prices_cleaned.xlsx", index=False)

df2
```
